```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold, GridSearchCV
```

```python
data = pd.read_csv("/content/drive/MyDrive/Cdac Project/Train_merge.csv")
data
```

```python
data.drop('Date', axis = 1, inplace=True)
data
```

```python
dum =  pd.get_dummies(data)
dum
```

```python
X = dum.drop('Weekly_Sales', axis = 1)
y = dum['Weekly_Sales']
```

```python
import warnings
warnings.filterwarnings('ignore')

kfold = KFold(n_splits=5, shuffle=True, random_state=2023)
cat_gbm =CatBoostRegressor(random_state=2023)

params ={'learning_rate': [0.03, 0.1],
    'depth':[4, 6, 10],
    'l2_leaf_reg': [1, 3, 5]}
    
gcv = GridSearchCV(cat_gbm, param_grid=params, cv=kfold)

gcv.fit(X, y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
