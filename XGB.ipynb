```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.pipeline import Pipeline

from sklearn.model_selection import KFold, GridSearchCV
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from xgboost import XGBRegressor
```

```python
data = pd.read_csv('/content/drive/MyDrive/Cdac Project/Train_merge2.csv', index_col=0)
data.drop('Date', axis = 1,  inplace=True)
```

```python
dum = pd.get_dummies(data)
dum
```

```python
X = dum.drop('Weekly_Sales', axis=1)
y = dum['Weekly_Sales']
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=2023)
xgb = XGBRegressor(random_state=2023)

params ={'learning_rate':[0.1, 0.15, 0.3, 0.35, 0.4, 0.5],
         'n_estimators':[25, 50, 75],
         'max_depth':[2,3,4,5]}
    
gcv = GridSearchCV(xgb, param_grid=params, cv=kfold)

gcv.fit(X, y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
